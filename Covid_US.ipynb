# Covid-19 USA- Visualization

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
 x= [1,2,3]
 y= [1,4,9]
plt.plot(x,y)
plt.title("Demo")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
sample = pd.read_csv('us-counties.csv')
sample.head(15)
```

```python


plt.figure(figsize=(13,10))
plt.title('Covid rate in USA')
plt.plot(sample.date,sample.cases, label ='cases')
plt.plot(sample.date,sample.deaths, label ='death', color='r')
#plt.plot(sample.date,sample.fips, label ='state')
plt.xlabel("Days")
plt.ylabel("No. of deaths and cases")




plt.legend()
plt.show()
```

```python

plt.plot(sample.county,sample.cases)
print(sample.date[1::10])
plt.show()
```

```python
samplee =pd.read_csv('us-counties.csv')
samplee = samplee[samplee.state == 'California']

plt.figure(figsize=(13,10))
plt.plot(samplee.date,samplee.cases, color='y', label ='case')
plt.plot(samplee.date,samplee.deaths, label ='death')
plt.xlabel("Days")
plt.ylabel("No. of deaths and cases")
plt.title("Covid Cases in California")
```

```python

samplee = samplee[samplee.state == 'California']
plt.figure(figsize=(8,5))
plt.ylabel('Number of range')
plt.xlabel('fips Level')
plt.title('Distribution of fips  in Covid-19 at California')

plt.hist(samplee.fips, color='#abcdef')
```
